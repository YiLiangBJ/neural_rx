```python
import tensorflow as tf

# 检查 GPU 是否可用
gpus = tf.config.list_physical_devices('GPU')
print(f"GPU 可用: {len(gpus) > 0}")
if gpus:
    # 获取 GPU 设备名称
    print(f"GPU 设备: {gpus[0].name}")
    # 打印 CUDA 和 cuDNN 版本（TensorFlow 不直接暴露 cuDNN 版本，需通过 tf.sysconfig）
    print(f"CUDA 版本: {tf.sysconfig.get_build_info()['cuda_version']}")
    print(f"cuDNN 版本: {tf.sysconfig.get_build_info()['cudnn_version']}")
else:
    print("未检测到 GPU")
```

```python
import torch

# 检查 GPU 是否可用
print(f"GPU 可用: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    # 获取 CUDA 设备名称
    print(f"GPU 设备: {torch.cuda.get_device_name(0)}")
    # 打印 CUDA 版本
    print(f"CUDA 版本: {torch.version.cuda}")
    # 打印 cuDNN 版本（需要 cuDNN 启用）
    print(f"cuDNN 版本: {torch.backends.cudnn.version() if torch.backends.cudnn.enabled else '未启用'}")
else:
    print("未检测到 GPU")
```
